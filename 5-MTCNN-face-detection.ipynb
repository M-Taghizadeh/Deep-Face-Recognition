```python
import mtcnn
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
print(mtcnn.__version__)
```

```python
# load image from file
pixels = plt.imread('./images/football_team.jpg')
```

```python
plt.imshow(pixels)
```

```python
detector = MTCNN()
# detect faces in the image
results = detector.detect_faces(pixels)
print(f"{len(results)} faces have been detected")
```

```python
results
```

```python
# draw an image with detected objects
def draw_facebox(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.figure(figsize=(15, 15))
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height, fill=False, color='red')
        # draw the box
        ax.add_patch(rect)
        # show the plot
    plt.show()
```

```python
draw_facebox('./images/football_team.jpg', results)
```

```python
# draw an image with detected objects
def draw_facebox_and_keypoints(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.figure(figsize=(15, 15))
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height,fill=False, color='red')
        # draw the box
        ax.add_patch(rect)
        # draw the dots
        for key, value in result['keypoints'].items():
            # create and draw dot
            dot = plt.Circle(value, radius=1, color='red')
            ax.add_patch(dot)
            # show the plot
    plt.show()# filename = 'test1.jpg' # filename is defined above, otherwise uncomment
```

```python
draw_facebox_and_keypoints('./images/football_team.jpg', results)
```

```python
# load image from file
pixels = plt.imread('./images/large_selfie.jpg')

detector = MTCNN()
# detect faces in the image
results = detector.detect_faces(pixels)
print(f"{len(results)} faces have been detected")

draw_facebox('./images/large_selfie.jpg', results)
```
